- NXDP1902103231 : 문어 200만 어절

```python
# load json file
with open('C:/big15/project_data/구문 분석 말뭉치(버전 2.0)/NXDP1902103231.json','r',encoding='utf-8-sig') as f:
    js = json.load(f)

# json -> dataframe
## document 내부에 data가 더 있음을 확인
js_data = pd.json_normalize(js)
js_data.head()
```

```python
# json -> dataframe
## document 풀기
js_document = pd.json_normalize(js['document'])
js_document
```

```python
# sentence 내부 구조 확인
## id, form, word, DP
print(type(js_document['sentence'][0]))
js_document['sentence'][0]
```

```python
# dataframe -> sentence 추출 -> DataFrame
## 이전 dataframe의 id를 가져올 필요가 있음
## 이전 id가 각 문장의 단어들을 구분
js_document_sentence = pd.DataFrame()
for i in tqdm(js_document.index) :
    sen_sentence = pd.json_normalize(js_document['sentence'][i])
    js_document_sentence = pd.concat([js_document_sentence, sen_sentence],ignore_index=True)
js_document_sentence
```

```python
# dataframe을 pickle로 저장 : js_document_sentence
with open("C:/dev/python-dev/AI-dev/workspace/data/corpus/parsing/js_document_sentence.pickle", "wb") as file:
    pickle.dump(js_document_sentence, file)
```

```python
# pickle을 dataframe으로 불러오기 : js_document_sentence
with open("C:/dev/python-dev/AI-dev/workspace/data/corpus/parsing/js_document_sentence.pickle", "rb" ) as file:
    js_document_sentence = pickle.load(file)
js_document_sentence
```

```python
# dataframe -> DP 추출 -> DataFrame
js_document_sentence_DP = pd.DataFrame()
for i in tqdm(js_document_sentence.index) :
    sen_sentence = pd.json_normalize(js_document_sentence['DP'][i])
    js_document_sentence_DP = pd.concat([js_document_sentence_DP, sen_sentence],ignore_index=True)
js_document_sentence_DP
```

```python
# dataframe을 json으로 변환 -> form을 examples를 풀 때도 가져가기 위함
test = js_document_sentence.to_json(orient='table')
parsed = json.loads(test)
parsed.keys()
```

```python
# js_document_sentence의 DP를 DataFrame 변환
parsing_clean = pd.json_normalize(data = parsed['data'], record_path='DP', meta=['id'], record_prefix='_')
print(parsing_clean.shape)
parsing_clean.head(10)
```
